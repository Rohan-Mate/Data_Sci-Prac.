```python
train = pd.read_csv('/kaggle/input/titanic/train.csv').drop(columns=['Survived'])
test = pd.read_csv('/kaggle/input/titanic/test.csv')
```

Single Sample t - test

```python
final_df = pd.concat([train,test]).sample(1309)
```

```python
pop = final_df['Age'].dropna()
```

```python
pop
```

```python
sample_age = pop.sample(25).values
```

```python
sample_age
```

```python
sample_age.mean()
```

```python
# H0 : the mean age is 35 
# H1 : the mean age is less than 35 
```

```python
# check the normality using shapiro wilk test 
from scipy.stats import shapiro

shapiro_age = shapiro(sample_age)

print(shapiro_age)
```

```python
pop_mean = 35 
```

```python
import scipy.stats as stats

t_statistic,p_value = stats.ttest_1samp(sample_age,pop_mean)

print('t_statistic:',t_statistic)
print('p_value:', p_value/2)
# we divide p_value by 2 bcoz of -> by default it gives two tail result 
# but in our condition -> one tail test
```

```python
alpha = 0.05 

if p_value < alpha:
    print('Reject the null hypothesis')
else:
    print('Fail to reject the null hypothesis')
```
